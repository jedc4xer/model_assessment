### XG Boost on Pima Native Data

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
# Import the data

diabetes_df = pd.read_csv('data/diabetes.csv')
diabetes_df
```

```python
dependent_vars = diabetes_df['Outcome'].copy()
independent_vars = diabetes_df.drop(columns = ['Outcome']).copy()
```

```python
X = independent_vars
y = dependent_vars
```

```python
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
dtrain = xgb.DMatrix(X_train, label = y_train)
dtest = xgb.DMatrix(X_test, label = y_test)

param = {
    'max_depth': 10,
    'eta': 0.3,
    'silent': 1
}

bst = xgb.train(param, dtrain, 20)
preds = bst.predict(dtest)

prediction_frame = pd.DataFrame(y_test, columns = ['ground_truth'])

prediction_frame['prediction'] = preds

from sklearn.metrics import r2_score

preds = [round(value) for value in preds]
accuracy = r2_score(y_test,preds)
```
